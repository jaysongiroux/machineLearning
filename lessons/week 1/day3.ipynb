# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# starter plot of x^2
x = np.arange(0,11)
y = x * x

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x^2')
plt.xlim(0,10)
plt.ylim(0,100)

# save the plot as a png
plt.savefig('x^2.png')
```

```python
fig = plt.figure()

# 
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('x^2')

# 1st number: distance from the left of the figure
# 2nd number: distance from the bottom of the figure
# 3rd number: proportion width of the figure
# 4th number: proportion height of the figure
zoomed_axes = fig.add_axes([.25,.5, .25, .25])
zoomed_axes.plot(x,y)
zoomed_axes.set_xlim([4,6])
zoomed_axes.set_ylim([15,40])
zoomed_axes.set_title("Zoomed in")
zoomed_axes.set_xlabel("x")
zoomed_axes.set_ylabel("y")
```

```python
a = np.arange(0,11)
b = a * 3

fig = plt.figure(dpi=100)
ax = fig.add_axes([0,0,1,1])
# exponential
ax.plot(x,y, label='Exponential')
# linear
ax.plot(a,b, label='Linear')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('x^2 vs x * 3')

# add legend
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y)
ax[1].plot(a, b)

fig.subplots_adjust(wspace=0.5)
```
